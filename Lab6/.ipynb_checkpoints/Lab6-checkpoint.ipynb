## Список столбцов файла `default_of_credit_card_clients.csv` (исходные данные)

---

- LIMIT_BAL – размер выданного кредита (новых тайваньских долларов);
- SEX – пол: 1 = мужской, 2 = женский;
- EDUCATION – образование: 1 = аспирантура, 2 = университет, 3 = старшая школа, 4 = другое;
- MARRIAGE – семейное положение: 1 = женат/замужем, 2 = одинок, 3 = другое;
- AGE – возраст, лет;
- PAY_0, PAY_2, ..., PAY_6 – статус последних ежемесячных платежей, с сентября (PAY_0) по апрель (PAY_6) 2005 года: -1 = платёж внесён вовремя, 1 = платёж внесён с опозданием на 1 месяц; 2 = с опозданием на 2 месяца, и т.д., 9 = с опозданием на 9 месяцев и более;
- BILL_AMT1, ..., BILL_AMT6 – размер последних ежемесячных платежей (новых тайваньских долларов), с сентября (BILL_AMT1) по апрель (BILL_AMT6) 2005 года;
- PAY_AMT1, ..., PAY_AMT6 – размер ежемесячных платежей в предыдущем месяце (новых тайваньских долларов), с сентября (PAY_AMT1 за август) по апрель (PAY_AMT6 за март) 2005 года;
- Y – целевая переменная: дефолт в следующем месяце (1 – да, 0 – нет).

---

| Номер варианта | Данные | Зависимая перемменная | Объясняющие переменные |
|:--------------:|:------|:---------------------:|:----------------------|
| 15             | defoult_of_credit <br> card_clients | -Y | все, кроме: Y, PAY_4, PAY_5, PAY_6, BILL_AMT4, BILL_AMT5, BILL_AMT6, PAY_AMT4, PAY_AMT5, PAY_AMT6 |

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder, OneHotEncoder
import time
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.model_selection import KFold, GridSearchCV
from sklearn.metrics import classification_report
```

```python
my_seed = 15
plt = mpl.pyplot
mpl.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("Set2")
```

```python
DF_raw = pd.read_csv('https://raw.githubusercontent.com/aksyuk/MTML/refs/heads/main/Labs/data/default_of_credit_card_clients.csv')
DF_raw = DF_raw.drop(['PAY_4', 'PAY_5', 'PAY_6', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6'], axis=1)

print('Число строк и столбцов в наборе данных:\n', DF_raw.shape)
```

```python
DF_raw.dtypes
```

```python
DF_raw['Y'].value_counts(dropna=False)
```

```python
# Числовые переменные
num_vars = ['LIMIT_BAL', 'AGE'] + [col for col in DF_raw.columns if any(x in col for x in ['BILL_', 'PAY_AMT'])]
# Качественные переменные
cat_vars = [col for col in DF_raw.columns if col not in num_vars]
```

```python
nom_col_names = ['SEX', 'EDUCATION', 'MARRIAGE']

one_hot = OneHotEncoder()

recoded = one_hot.fit_transform(DF_raw[nom_col_names]).toarray()

clmns = one_hot.get_feature_names_out(nom_col_names)
df_dummy_nom = pd.DataFrame(recoded, columns=clmns)

print(df_dummy_nom.shape)

df_dummy_nom.head()
```

```python
ord_cat_vars = []
for col in cat_vars:
	if col not in nom_col_names:
		ord_cat_vars.append(col)
```

```python
DF_raw[ord_cat_vars].shape
```

```python
DF = pd.concat([DF_raw[num_vars].reset_index(), DF_raw[ord_cat_vars].reset_index(), df_dummy_nom], axis=1)

print('Размерность обучающейго фрейма после перекодировки:', DF.shape)

DF = DF.drop('index', axis=1)

DF.head()
```

```python
DF.dtypes
```

```python
DF_teach = DF.sample(frac=0.85, random_state=my_seed)
DF_predict = DF.drop(DF_teach.index)
```

##  Преобразование исходных данных и построение моделей

```python
DF_teach[num_vars].describe()
```

### Стандартизация и переход к главным компонентам

```python
sc = StandardScaler()
X_train_std = sc.fit_transform(DF_teach.drop('Y', axis=1).values)
pca = PCA()
X_train_pca = pca.fit_transform(X_train_std)
frac_var_expl = pca.explained_variance_ratio_
print('Доли объяснённой дисперсии по компонентам в PLS:\n', np.around(frac_var_expl, 3), '\nОбщая сумма долей:', np.around(sum(frac_var_expl), 3))
```

## Модель логистической регрессии с перекрестной проверкой

Для svm с перекрестной проверкой используется только выборка составляющая 1% от тренировочной, что составляет 255 наблюдений. В противном случае время обработки превышает 200min и не завершается.

```python
X_train = DF_teach.sample(frac=0.01, random_state=my_seed).drop('Y', axis=1)
y_train = DF_teach.sample(frac=0.01, random_state=my_seed)['Y']

pipe_lr = make_pipeline(StandardScaler(), PCA(n_components=3), LogisticRegression(random_state=my_seed, solver='lbfgs'))
# будем сохранять точность моделей в один массив:
score = list()
score_models = list()
# считаем точность с перекрёстной проверкой, показатель Acc
cv = cross_val_score(estimator=pipe_lr, X=X_train, y=y_train, cv=5, scoring='accuracy')
# записываем точность
score.append(np.around(np.mean(cv), 3))
score_models.append('sc_pca_logit')
print('Acc с перекрёстной проверкой', '\nдля модели', score_models[0], ':', score[0])
```

```python
X_train.shape
```

```python
pipe_svc = make_pipeline(StandardScaler(), SVC(random_state=my_seed))

param_range = [0.00001, 0.0001, 0.001, 0.01, 0.1, 1.0, 10.0, 100.0, 1000.0, 10000.0]
param_grid = [{'svc__C': param_range,
	'svc__kernel': ['linear']},
	{'svc__C': param_range,
	'svc__gamma': ['scale', 'auto', 0.001, 0.01, 0.1, 1.0],
	'svc__kernel': ['rbf'],
	'svc__class_weight': [None, 'balanced']},
	{'svc__C': param_range,
	'svc__gamma': param_range,
	'svc__degree' : [2, 3],
	'svc__kernel': ['poly']}]

kfold = KFold(n_splits=5, random_state=my_seed, shuffle=True)
gs = GridSearchCV(estimator=pipe_svc, param_grid=param_grid, scoring='accuracy', refit=True, cv=kfold, n_jobs=-1)

tic = time.perf_counter()

gs = gs.fit(X_train, y_train)

toc = time.perf_counter()
print(f"Сеточный поиск занял {toc - tic:0.2f} секунд", sep='')
```

```python
np.around(gs.best_score_, 3)
```

```python
gs.best_estimator_.get_params()['svc__kernel']
```

```python
gs.best_estimator_.get_params()['svc__C']
```

```python
score.append(np.around(gs.best_score_, 3))
score_models.append('sc_pca_svc')
print('Acc с перекрёстной проверкой', '\nдля модели', score_models[1], ':', score[1])
```

## Прогноз по лучшей модели

```python
pd.DataFrame({'Модель' : score_models, 'Acc' : score})
```

```python
y_hat = gs.best_estimator_.predict(X=DF_predict.drop('Y', axis=1))
print(classification_report(DF_predict['Y'], y_hat))
```
